## Goal
Inspect the raw grocery dataset to understand its structure, columns, and quality.

## Steps
1. Load the CSV.
2. Inspect shape, columns, and data types.
3. Check for missing values.
4. Summarize key statistics.

```python
import pandas as pd
import os

# 1. Load the CSV
DATA_PATH = "raw/canada_grocery_nutrition_5000.csv"

if not os.path.exists(DATA_PATH):
    print(f"Error: File not found at {DATA_PATH}")
else:
    df = pd.read_csv(DATA_PATH)
    print("Dataset loaded successfully!")
```

```python
# 2. Basic Inspection
print("\n--- Head ---")
display(df.head())

print("\n--- Shape ---")
print(df.shape)

print("\n--- Info ---")
print(df.info())

print("\n--- Columns ---")
print(df.columns.tolist())
```

```python
# 3. Statistical Summary
print("\n--- Describe (Numerical) ---")
display(df.describe().T)
```

```python
# 4. Missing Values
print("\n--- Missing Values ---")
missing = df.isnull().sum()
print(missing[missing > 0])
```
